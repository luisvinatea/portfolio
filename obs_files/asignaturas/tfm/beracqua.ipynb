## Which type of Aerator should we invest in?

To assess the viability of investing in aquaculture farming equipment based on the price findings for diesel and electric aerators, we can apply several financial methods such as **Discounted Cash Flow (DCF)**, **Net Present Value (NPV)**, and **Internal Rate of Return (IRR)**. These methods will help determine whether the short-term cost savings of cheaper Chinese diesel aerators offset their lower durability compared to premium aerators like those provided by **BERAQUA**.

Here’s a step-by-step approach to assess this viability:

### 1. **Key Variables to Consider**:
   - **Initial Investment**: The upfront cost of purchasing the aerators.
     - Chinese Diesel Aerators: Typically lower-priced.
     - BERAQUA Electric Aerators: Higher upfront cost.
   - **Durability** (useful life of aerators):
     - Chinese Diesel Aerators: Lower durability.
     - BERAQUA Electric Aerators: Higher durability (assumed to last longer).
   - **Operating Costs**: 
     - Diesel-powered aerators typically have higher fuel costs and maintenance expenses.
     - Electric aerators have lower fuel costs but may require more initial infrastructure investment.
   - **Depreciation**: The method of depreciating the cost of the aerators over their useful life.
   - **Cash Flow Impact**: The capital saved or spent on repairs, maintenance, and replacement over time.
   - **Discount Rate**: The rate at which future cash flows will be discounted. This reflects the cost of capital or expected return on investment.
   - **Net Present Value (NPV)** and **Internal Rate of Return (IRR)**: These will be the primary metrics to determine if the investment is worthwhile.

### 2. **Steps to Calculate Viability**:

#### a. **Estimate Initial Investment**:
- Use the prices we have for **Chinese Diesel Aerators** and **BERAQUA Electric Aerators**.
  - Average price for Chinese Diesel Aerators: $589.30 (from the stats provided).
  - Average price for BERAQUA Electric Aerators (we’ll assume higher, possibly $1,500 based on other electric aerators’ prices).

#### b. **Estimate Operating Costs and Durability**:
- **Operating Costs**:
  - Diesel Aerators: Higher ongoing costs due to fuel consumption and frequent maintenance.
  - Electric Aerators: Lower operating costs, assuming stable electricity prices.
  
- **Durability**:
  - Chinese Diesel Aerators: Assume a lifespan of 3-5 years (shorter).
  - BERAQUA Electric Aerators: Assume a lifespan of 10-15 years (longer).

#### c. **Discounted Cash Flow (DCF) Model**:
This will account for the time value of money and will discount future cash flows (savings or expenses) to present value using a discount rate.

$ \text{DCF} = \sum_{t=1}^{n} \frac{C_t}{(1 + r)^t} $

Where:
- $ C_t $ is the net cash flow in year $ t $ (revenue - costs),
- $ r $ is the discount rate,
- $ n $ is the number of years (durability).

#### d. **Depreciation**:
Apply straight-line depreciation to the initial investment cost over the lifespan of each aerator:
$ \text{Depreciation per year} = \frac{\text{Initial Cost}}{\text{Lifespan}} $

#### e. **Net Present Value (NPV)**:
Calculate the NPV of both types of aerators using the DCF formula. The NPV will help us decide if the investment in **BERAQUA Electric Aerators** is more favorable despite the higher initial cost.

#### f. **Internal Rate of Return (IRR)**:
Calculate the IRR for each aerator type to determine the rate of return that sets the NPV to zero. A higher IRR indicates a better investment.

- **Replacement costs** (for Chinese Diesel Aerators, given their shorter lifespan).

### Additional Financial Metrics and Insights:

1. **Payback Period**:
   - The **Payback Period** tells us how long it takes to recover the initial investment from cash inflows (revenue or savings).
   - **Formula**:
     $$
     \text{Payback Period} = \frac{\text{Initial Investment}}{\text{Annual Net Cash Flow (Revenue - Operating Costs)}}
     $$

2. **Profitability Index (PI)**:
   - The **Profitability Index** is a ratio of the present value of future cash flows to the initial investment. It helps in deciding if an investment is profitable.
   - **Formula**:
     $$
     \text{PI} = \frac{\text{NPV} + \text{Initial Investment}}{\text{Initial Investment}}
     $$
     A PI greater than 1 indicates a profitable investment.

3. **Sensitivity Analysis**:
   - You can perform a **sensitivity analysis** by varying key parameters such as:
     - Freight costs
     - Operating costs
     - Revenue
     - Discount rate
   - This will help in evaluating the impact of changing conditions on the NPV and IRR.

4. **Freight Impact on Total Cost**:
   - Knowing the **freight costs** brings additional clarity to the total import expenses and helps in comparing alternative models from different suppliers.
   - For instance, if one supplier has a higher CIF but lower freight costs, it might still be a better option in terms of the total import cost, depending on how much freight contributes to the total.

   **Additional insights include**:
   - **Comparing suppliers** based on FOB and freight costs, identifying those that offer better deals.
   - **Analyzing geographical trends**, such as how freight costs vary by the country of origin.

```python
# Function to calculate DCF, NPV, IRR, Payback Period, and Profitability Index
def calculate_metrics(data):
    # Extract relevant data
    initial_investment = data['initial_investment']
    operating_costs_per_year = data['operating_costs_per_year']
    revenue_per_year = data['revenue_per_year']
    lifespan = data['lifespan']
    discount_rate = data['discount_rate']
    
    # Generate cash flows (revenue - operating costs)
    cash_flows = [-initial_investment] + [(revenue_per_year - operating_costs_per_year) for _ in range(lifespan)]
    
    # Calculate NPV and IRR
    npv = npf.npv(discount_rate, cash_flows)
    irr = npf.irr(cash_flows)
    
    # Calculate Discounted Cash Flows (DCF)
    dcf = [cf / (1 + discount_rate) ** i for i, cf in enumerate(cash_flows)]
    
    # Calculate Payback Period (Time to recover initial investment)
    cumulative_cash_flow = 0
    payback_period = 0
    for i, cf in enumerate(cash_flows[1:], start=1):  # Start from year 1
        cumulative_cash_flow += cf
        if cumulative_cash_flow >= initial_investment:
            payback_period = i
            break
    
    # Calculate Profitability Index (PI)
    pi = (npv + initial_investment) / initial_investment
    
    return {
        "cash_flows": cash_flows,
        "dcf": dcf,
        "npv": npv,
        "irr": irr,
        "payback_period": payback_period,
        "profitability_index": pi
    }

# Data for AQUA B-309 with assumed revenue
aqua_b_309 = {
    "initial_investment": 1268.47,  # Initial investment (e.g., CIF)
    "operating_costs_per_year": 877.32,  # Annual operating costs (energy + maintenance)
    "revenue_per_year": 1000,  # Annual revenue or savings
    "lifespan": 10,  # Estimated lifespan in years
    "discount_rate": 0.10,  # 10% discount rate
}

# Perform the calculation for AQUA B-309
results = calculate_metrics(aqua_b_309)

# Display the results
print(f"Cash Flows: {results['cash_flows']}")
print(f"Discounted Cash Flows: {results['dcf']}")
print(f"Net Present Value (NPV): {results['npv']:.2f}")
print(f"Internal Rate of Return (IRR): {results['irr']:.2%}")
print(f"Payback Period: {results['payback_period']} years")
print(f"Profitability Index (PI): {results['profitability_index']:.2f}")
```
