## Which type of Aerator should we invest in?

To assess the viability of investing in aquaculture farming equipment based on the price findings for diesel and electric aerators, we can apply several financial methods such as **Discounted Cash Flow (DCF)**, **Net Present Value (NPV)**, and **Internal Rate of Return (IRR)**. These methods will help determine whether the short-term cost savings of cheaper Chinese diesel aerators offset their lower durability compared to premium aerators like those provided by **BERAQUA**.

Here’s a step-by-step approach to assess this viability:

### 1. **Key Variables to Consider**:
   - **Initial Investment**: The upfront cost of purchasing the aerators.
     - Diesel Aerators: Typically lower-priced.
     - BERAQUA Electric Aerators: Higher upfront cost.
   - **Durability** (useful life of aerators):
     - Chinese Diesel Aerators: Lower durability.
     - BERAQUA Electric Aerators: Higher durability (assumed to last longer).
   - **Operating Costs**: 
     - Diesel-powered aerators typically have higher fuel costs and maintenance expenses.
     - Electric aerators have lower fuel costs but may require more initial infrastructure investment.
   - **Depreciation**: The method of depreciating the cost of the aerators over their useful life.
   - **Cash Flow Impact**: The capital saved or spent on repairs, maintenance, and replacement over time.
   - **Discount Rate**: The rate at which future cash flows will be discounted. This reflects the cost of capital or expected return on investment.
   - **Net Present Value (NPV)** and **Internal Rate of Return (IRR)**: These will be the primary metrics to determine if the investment is worthwhile.

### 2. **Steps to Calculate Viability**:

#### a. **Estimate Initial Investment**:
- Use the prices we have for **Chinese Diesel Aerators** and **BERAQUA Electric Aerators**.
  - Average price for Chinese Diesel Aerators: $589.30 (from the stats provided).
  - Average price for BERAQUA Electric Aerators (we’ll assume higher, possibly $1,500 based on other electric aerators’ prices).

#### b. **Estimate Operating Costs and Durability**:
- **Operating Costs**:
  - Diesel Aerators: Higher ongoing costs due to fuel consumption and frequent maintenance.
  - Electric Aerators: Lower operating costs, assuming stable electricity prices.
  
- **Durability**:
  - Chinese Diesel Aerators: Assume a lifespan of 2-3 years (shorter).
  - BERAQUA Electric Aerators: Assume a lifespan of 5-6 years (longer).

#### c. **Discounted Cash Flow (DCF) Model**:
This will account for the time value of money and will discount future cash flows (savings or expenses) to present value using a discount rate.

$ \text{DCF} = \sum_{t=1}^{n} \frac{C_t}{(1 + r)^t} $

Where:
- $ C_t $ is the net cash flow in year $ t $ (revenue - costs),
- $ r $ is the discount rate,
- $ n $ is the number of years (durability).

#### d. **Depreciation**:
Apply straight-line depreciation to the initial investment cost over the lifespan of each aerator:
$ \text{Depreciation per year} = \frac{\text{Initial Cost}}{\text{Lifespan}} $

#### e. **Net Present Value (NPV)**:
Calculate the NPV of both types of aerators using the DCF formula. The NPV will help us decide if the investment in **BERAQUA Electric Aerators** is more favorable despite the higher initial cost.

#### f. **Internal Rate of Return (IRR)**:
Calculate the IRR for each aerator type to determine the rate of return that sets the NPV to zero. A higher IRR indicates a better investment.


### Additional Financial Metrics and Insights:

1. **Payback Period**:
   - The **Payback Period** tells us how long it takes to recover the initial investment from cash inflows (revenue or savings).
   - **Formula**:
     $$
     \text{Payback Period} = \frac{\text{Initial Investment}}{\text{Annual Net Cash Flow (Revenue - Operating Costs)}}
     $$

2. **Profitability Index (PI)**:
   - The **Profitability Index** is a ratio of the present value of future cash flows to the initial investment. It helps in deciding if an investment is profitable.
   - **Formula**:
     $$
     \text{PI} = \frac{\text{NPV} + \text{Initial Investment}}{\text{Initial Investment}}
     $$
     A PI greater than 1 indicates a profitable investment.

```python
import pandas as pd
import numpy_financial as npf

# Aerator data with detailed costs and performance metrics (including Chinese aerator)
data1 = {
    "AQUA B-309": {
        "Initial Investment": 964 + 262 + 51.59 + 12.88,  # Summing all cost components
        "Annual Energy Cost": 577.39,
        "Annual Maintenance Cost": 186.75,
        "Annual Replacement Cost": 13.18,
        "Lifespan": 5,  # years
        "SOTR": 1.53,
        "SAE": 1.38
    },
    "SAA-308": {
        "Initial Investment": 1126.89 + 66.53 + 12.88 + 30,
        "Annual Energy Cost": 682.98,
        "Annual Maintenance Cost": 369.32,
        "Annual Replacement Cost": 75.00,
        "Lifespan": 4,
        "SOTR": 1.42,
        "SAE": 0.77
    },
    "PA-308": {
        "Initial Investment": 1350.48 + 66.79 + 12.88 + 30,
        "Annual Energy Cost": 682.98,
        "Annual Maintenance Cost": 369.32,
        "Annual Replacement Cost": 13.18,
        "Lifespan": 4,
        "SOTR": 1.42,
        "SAE": 1.13
    },
    "Chinese Aerator": {
        "Initial Investment": 700,  # From the data provided
        "Annual Energy Cost": 671.81,
        "Annual Maintenance Cost": 369.32,
        "Annual Replacement Cost": 75.00,
        "Lifespan": 2,  # Lifespan of the Chinese aerator
        "SOTR": 1.0,  # Assuming lower SOTR due to 100 RPM
        "SAE": 0.8  # Assumed lower efficiency due to the lower RPM
    },
    "OXIPEZ": {
        "Initial Investment": 828.49,
        "Annual Energy Cost": 0.00,  # Missing data filled as 0 based on provided input
        "Annual Maintenance Cost": 363.37,
        "Annual Replacement Cost": 0.00,  # Assuming zero based on missing data
        "Lifespan": 5,  # Provided
        "SOTR": 1.50,
        "SAE": 1.00  # Estimated SAE
    }
}

# Constants from the image data
yield_15_percent = 2976  # kg/ha at 15% saturation
yield_40_percent = 3631  # kg/ha at 40% saturation
yield_65_percent = 3975  # kg/ha at 65% saturation

# Interpolating yield at 50% saturation
yield_50_percent = yield_40_percent + ((50 - 40) / (65 - 40)) * (yield_65_percent - yield_40_percent)
print(f"Estimated yield at 50% oxygen saturation: {yield_50_percent:.2f} kg/ha")

# Aerator data from data_sheet8 (converted to usable format, adding Chinese Aerator)
data2 = {
    "Aerator_Model": ["DIVA", "ECUASINO", "CRUPESA", "BERAQUA - AQUAMIX", "OXIPEZ", "Chinese Aerator"],
    "HP": [2, 2, 2.5, 1.5, 2, 2],  # Assuming 2 HP for the Chinese Aerator
    "Capital_Investment_USD": [492.80, 492.80, 450.00, 797.62, 828.49, 700.00],
    "Operating_Cost_USD_per_ha": [2064.10, 1904.29, 1961.39, 1133.67, 1000, 2241.13],  # Estimate operating cost for OXIPEZ
    "Lifespan_Years": [4, 4, 4, 5, 5, 2],  # Lifespan of Chinese Aerator is 2 years
    "RPM": [125, 100, 125, 125, 125, 100],  # Chinese Aerator has 100 RPM
}

# Create DataFrame for the second dataset
df_aerators = pd.DataFrame(data2)

# Constants
price_per_kg_shrimp = 4.50
fcr = 2.0  # Feed conversion ratio
feed_price_per_kg = 1.2  # USD per kg of feed
scale = 100  # Size of operation in hectares
discount_rate = 0.10  # 10% discount rate

# Calculate total operating cost over the lifespan of the aerator for the first dataset
def calculate_total_cost(model_data):
    initial_investment = model_data['Initial Investment']
    total_operating_cost = (model_data['Annual Energy Cost'] + 
                            model_data['Annual Maintenance Cost'] + 
                            model_data['Annual Replacement Cost']) * model_data['Lifespan']
    total_cost_of_ownership = initial_investment + total_operating_cost
    return total_cost_of_ownership

# Function to handle missing data and fallback to default values
def safe_get(data, key, default):
    return data.get(key, default)

# Iterate over aerator models to calculate profitability for the first dataset
for model, model_data in data1.items():
    # Handle missing SAE by defaulting to 1.0 if it's not present
    sae = safe_get(model_data, 'SAE', 1.0)
    
    # Adjust total yield by the aerator's performance (e.g., using SOTR or SAE)
    total_yield = yield_50_percent * scale * sae  # Adjusted by SAE
    
    total_revenue = total_yield * price_per_kg_shrimp
    total_cost = calculate_total_cost(model_data)
    
    print(f"Aerator Model: {model}")
    print(f"  Total Yield (kg/year) at 50% oxygen saturation: {total_yield:.2f}")
    print(f"  Total Revenue (USD/year): {total_revenue:.2f}")
    print(f"  Total Cost of Ownership over {model_data['Lifespan']} years: {total_cost:.2f}")
    print("-" * 50)

# Function to calculate financial metrics for the second dataset, including corrected payback period
def calculate_metrics(row):
    # Using interpolated yield at 50% oxygen saturation
    annual_yield = yield_50_percent * row['HP'] * scale
    annual_revenue = annual_yield * price_per_kg_shrimp
    
    # Feed costs
    annual_feed_cost = annual_yield * fcr * feed_price_per_kg
    
    # Operating costs
    operating_cost = row["Operating_Cost_USD_per_ha"] * scale
    
    # Net cash flow
    net_cash_flow = annual_revenue - annual_feed_cost - operating_cost
    
    # Cash flows
    initial_investment = -row['Capital_Investment_USD'] * scale
    cash_flows = [initial_investment] + [net_cash_flow] * row['Lifespan_Years']
    
    # NPV and IRR
    npv = npf.npv(discount_rate, cash_flows)
    irr = npf.irr(cash_flows)
    
    # Corrected Payback Period Calculation
    cumulative_cash_flow = 0
    payback_period = None
    
    for i, cf in enumerate(cash_flows[1:], start=1):  # Start with year 1 (index 1)
        cumulative_cash_flow += cf
        if cumulative_cash_flow >= -initial_investment:
            payback_period = i
            break
    
    return pd.Series([npv, irr, payback_period])

# Apply metrics to DataFrame for the second dataset
df_aerators[['NPV', 'IRR', 'Payback Period']] = df_aerators.apply(calculate_metrics, axis=1)

# Format the NPV as currency (without scientific notation)
df_aerators['NPV'] = df_aerators['NPV'].apply(lambda x: f"${x:,.2f}")

# Display results for the second dataset
print(df_aerators)

```

```python
import matplotlib.pyplot as plt  # type: ignore
from sklearn.preprocessing import PolynomialFeatures # type: ignore
from sklearn.linear_model import LinearRegression # type: ignore
from mpl_toolkits.mplot3d import Axes3D # type: ignore

# Sample data for productivity, FCR, and NPV
data = {
    "Productivity (kg/ha)": [3000, 3250, 3500, 3750, 4000, 4250, 4500, 4750],
    "FCR": [1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 1.7],
    "NPV": [5.355844e+06, 5.244395e+06, 5.132946e+06, 5.026046e+06, 4.915597e+06, 4.805148e+06, 4.694699e+06, 4.584250e+06]
}

df = pd.DataFrame(data)

# Prepare the data for quadratic regression (use values as NumPy arrays)
X = df[["Productivity (kg/ha)", "FCR"]].values
y = df["NPV"].values

# Create polynomial features (quadratic terms)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the quadratic model
model = LinearRegression()
model.fit(X_poly, y)

# Predict using the model
y_pred = model.predict(X_poly)

# Create a 3D plot for visualization
fig = plt.figure(figsize=(36, 24))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for the actual data with larger markers and different color
ax.scatter(df["Productivity (kg/ha)"], df["FCR"], df["NPV"], c='r', marker='o', s=50, label='Actual Data')

# Plot the predicted quadratic surface
x = np.linspace(df["Productivity (kg/ha)"].min(), df["Productivity (kg/ha)"].max(), 100)
fcr = np.linspace(df["FCR"].min(), df["FCR"].max(), 100)
X_grid, FCR_grid = np.meshgrid(x, fcr)
Z = model.predict(poly.transform(np.c_[X_grid.ravel(), FCR_grid.ravel()])).reshape(X_grid.shape)

# Plot the surface with transparency and a coolwarm colormap
ax.plot_surface(X_grid, FCR_grid, Z, alpha=0.6, cmap='coolwarm', edgecolor='none')

# Set axis labels and title
ax.set_xlabel('Productivity (kg/ha)', fontsize=12)
ax.set_ylabel('FCR', fontsize=12)
ax.set_zlabel('NPV (USD)', fontsize=12)
ax.set_title('Quadratic Model for NPV vs Productivity and FCR', fontsize=14)

# Show legend for data points
ax.legend()

plt.show()


```

```python
from scipy.optimize import minimize

# Define a function to calculate the NPV given key variables: FCR, Aerator Power
def calculate_npv(params, price_per_kg, feed_cost, production_scale, energy_cost, maintenance_cost, capital_investment, lifespan, discount_rate, chinese=False):
    fcr, aerator_hp = params
    oxygen_saturation = 0.50  # Fixed at 50% for all aerators
    
    if chinese:
        # Chinese aerator has reduced efficiency and shorter lifespan
        oxygen_saturation = 0.40  # Reduced due to lower RPM efficiency (100 RPM)
        lifespan = 2  # Chinese aerator lifespan is 2 years instead of 5 or 6
    
    # Calculate annual yield based on production scale, oxygen saturation, and aerator horsepower
    yield_per_hp = 500  # kg/HP (adjust as needed for aerator efficiency)
    annual_yield = yield_per_hp * aerator_hp * production_scale * (52 / 16)  # 16-week growth cycle
    
    # Calculate revenue
    revenue = annual_yield * price_per_kg
    
    # Calculate feed costs
    feed_cost_total = annual_yield * fcr * feed_cost
    
    # Total operating costs (energy + maintenance)
    total_operating_costs = energy_cost * aerator_hp * production_scale + maintenance_cost
    
    # Calculate total cost (initial + operating costs over lifespan)
    total_costs = capital_investment + (total_operating_costs * lifespan)
    
    # Net cash flow (revenue - costs)
    net_cash_flow = revenue - feed_cost_total - total_operating_costs
    
    # Discounted cash flows (DCF)
    discounted_cash_flows = [net_cash_flow / ((1 + discount_rate) ** year) for year in range(1, lifespan + 1)]
    
    # Net Present Value (NPV)
    npv = sum(discounted_cash_flows) - capital_investment
    
    return -npv  # Minimize NPV for optimization purposes

# Initial guess for the parameters: FCR, Aerator HP
initial_guess = [1.5, 2.0]  # FCR 1.5, Aerator HP 2.0

# Constraints for FCR (1.2 - 2.5) and aerator horsepower (1.0 - 3.0)
bounds = [(1.2, 2.5), (1.0, 3.0)]

# Optimization for a standard aerator (non-Chinese)
result_standard = minimize(
    calculate_npv,
    initial_guess,
    args=(4.5, 1.5, 100, 0.06, 186.75, 1200, 5, 0.1),  # price_per_kg, feed_cost, production_scale, energy_cost, maintenance_cost, capital_investment, lifespan, discount_rate
    bounds=bounds,
    method='SLSQP'
)

# Optimization for the Chinese aerator
result_chinese = minimize(
    calculate_npv,
    initial_guess,
    args=(4.5, 1.5, 100, 0.06, 186.75, 1200, 5, 0.1, True),  # same parameters but with 'chinese=True' flag
    bounds=bounds,
    method='SLSQP'
)

# Optimal parameters after optimization
optimal_fcr_standard, optimal_aerator_hp_standard = result_standard.x
optimal_fcr_chinese, optimal_aerator_hp_chinese = result_chinese.x

# Optimal NPV calculations
optimal_npv_standard = -result_standard.fun
optimal_npv_chinese = -result_chinese.fun

# Output results
print("Optimal FCR (Standard Aerator):", optimal_fcr_standard)
print("Optimal Aerator HP (Standard Aerator):", optimal_aerator_hp_standard)
print("Optimal NPV (Standard Aerator):", optimal_npv_standard)

print("\nOptimal FCR (Chinese Aerator):", optimal_fcr_chinese)
print("Optimal Aerator HP (Chinese Aerator):", optimal_aerator_hp_chinese)
print("Optimal NPV (Chinese Aerator):", optimal_npv_chinese)
```
