# Los Elementos de la Decisión de Inversión

1. **La inversión requerida o inversión inicial**: Parte de la idea inicial y pasa por la compra de los activos hasta el lanzamiento del producto.
2. **La duración de la inversión**: En términos de plazos a cumplir.
3. **Los ingresos y pagos derivados del proyecto**: El flujo neto de caja se obtiene por diferencia entre las entradas y las salidas durante la vida del proyecto.

## Inversión Inicial

1. La inversión inicial es el total de fondos que se requieren para iniciar el proyecto. Esto incluye el coste económico de los activos y todos los demás conceptos necesarios para la implementación del proyecto.
2. La inversión inicial es la suma de las inversiones en activos fijos más la tesorería que se necesita para iniciar las operaciones.
3. Desde un punto de vista práctico, el capital invertido se considera una salida neta en el momento 0, y se concreta en la suma actualizada de todos los pagos a incluir en este concepto.

## Horizonte Temporal

El horizonte temporal, o vida del proyecto, se define como el tiempo en el que se materializa la inversión (desde el momento en que se produce la primera obligación de pago) y hasta que se para el proceso de generación de ingresos. Asumir un determinado horizonte temporal no significa que el proyecto se limite a ese período, sino que se delimita el período a efectos de cálculo.

## Flujo Libre de Caja

El flujo libre de caja representa el flujo de caja de la empresa y viene dado por la diferencia entre los flujos de caja operacional y el de las inversiones en activo no corriente. No tiene en cuenta la financiación, pues se centra en el coste real del activo sin estar afectado por una financiación económica. El efecto de la financiación se tiene en cuenta en el cálculo del WACC.

### Flujo de Caja Libre Operativo

Se obtiene de la diferencia entre el Flujo de Caja Bruto (FCB) y la Inversión Bruta (IB).

\begin{align*}
  \text{FCB} &= \text{BAIT} - (t \times \text{BAIT}) = \text{BAIDT} \, (\text{Beneficio Antes de Intereses Después de Impuestos}) + \text{Amortizaciones}
\end{align*}

### Inversión Bruta (IB)

Hay que diferenciar entre la inversión en Activo Fijo (Activo No Corriente) y la realizada en el Fondo de Maniobra.


$\text{Inversion en Activo Fijo} = \text{AFN}(t) - \text{AFN}(t-1) + \text{Amortizaciones}(t)$\

## Fondo de Maniobra Neto

Muchas inversiones requieren capital de trabajo adicional. Un aumento del FMN es una salida de efectivo, mientras que un descenso supone una entrada.

- **FMN =** Activo corriente - Pasivo corriente
- **FMN =** Inventario promedio + promedio de clientes - promedio de proveedores

## Flujo Libre de Caja


$\text{Margen Operativo Bruto (EBITDA)} = \text{Ventas} - \text{Coste Ventas} - \text{Gastos Generales}$\
$\text{BAIT} = \text{EBITDA} - \text{Amortizaciones}$\
$\text{BAIDT} = \text{BAIT} - \text{BAIT} \times (1-t)$\
$\text{FCB} = \text{BAIDT} + \text{Amortizaciones}$


### Inversión Bruta

$\text{Inversion en AF} = \text{AFN}(t) - \text{AFN}(t-1) + \text{Amortizaciones}(t)$\
$\text{Inversion en FM} = \text{NOF}(t) - \text{NOF}(t-1)$


### NOF (Necesidades Operativas de Financiación


$\text{NOF} = \text{Caja operativa} + \text{Clientes} + \text{Existencias} - \text{Proveedores}$\
$\text{FCF} = \text{FCB} - \text{IB}$


## Métodos Estáticos en la Decisión de Inversiones

1. Tener en cuenta el valor del dinero en el tiempo.
2. Tener en cuenta el riesgo.
3. Estar basado en el Flujo Libre de Caja.
4. Ordenar los proyectos de forma adecuada.
5. Tomar la decisión de inversión que maximice el valor de los accionistas.

### Plazo de Recuperación (Payback Period)

El plazo de recuperación es el tiempo que se requiere para que la empresa recupere su inversión inicial.

- Si el plazo de recuperación es menor que el plazo máximo aceptable fijado por la Dirección, se acepta el proyecto.
- Si el plazo de recuperación es mayor que el plazo máximo aceptable fijado por la Dirección, se rechaza el proyecto.

La Dirección establece el plazo de recuperación máximo aceptable para cada proyecto. El plazo máximo fijado por la dirección es de 2,75 años.

**Proyecto Western Europe:** 

Inversión inicial de \$250M  
Las entradas de efectivo en los primeros tres años son de sólo \$245 millones.  
- Global Wireless rechazará el proyecto (3 > 2,75).

**Proyecto Southeast U.S.:**

Inversión inicial de 50M  
- Los flujos de caja de los primeros 2 años acumulan 40 millones.
- El proyecto recupera la inversión inicial después de 2,40 años.
- Las entradas en el año 3 son de 25 millones, por lo que el proyecto genera 10 millones en el año 3 en tan sólo 0,40 años ($ \$10 ÷ \$25 $).
- Global Wireless aceptará el proyecto (2,4 < 2,75).

### Ventajas y Desventajas del Plazo de Recuperación

**Ventajas:**
- Fácil de calcular.
- Fácil de entender.
- Se basa en los flujos libres de caja.

**Desventajas:**
- No tiene en cuenta el valor del dinero en el tiempo.
- No tiene en cuenta el riesgo.
- El plazo máximo fijado por la Dirección es arbitrario.
- No maximiza el valor del accionista.

## Rentabilidad de los Flujos de Caja

La rentabilidad de los flujos de caja es el retorno de la inversión por cada euro invertido.

- Si la rentabilidad del flujo de caja es mayor a 1, aceptamos el proyecto.
- Cuanto mayor es la rentabilidad del flujo de caja, mejor.

**Proyecto Western Europe:**

$r_{WE} = \frac{35 + 80 + 130 + 160 + 175}{250} \approx 2,32 \, (232\%)$

**Proyecto Southeast U.S.:**

$r_{SE} = \frac{18 + 22 + 25 + 30 + 32}{50} \approx 2,54 \, (254\%)$

El proyecto Southeast U.S. tiene una mayor rentabilidad del flujo de caja.

### Ventajas y Desventajas del Método de Rentabilidad de los FC

**Ventajas:**
- Su cálculo es muy sencillo.
- Es fácil de interpretar.
- Se basa en los flujos libres de caja.

**Desventajas:**
- No tiene en cuenta el valor del dinero en el tiempo.
- No tiene en cuenta el riesgo.
- La rentabilidad obtenida no es anual, sino la acumulada durante toda la vida del proyecto.

## Rentabilidad Anual Promedio de los Flujos de Caja por Euro Invertido

La rentabilidad de los flujos de caja es el promedio anual de los flujos de caja, por euro invertido.

- Si la rentabilidad del proyecto es mayor que 1, aceptamos el proyecto.
- Elegiremos el proyecto de mayor rentabilidad.

**Proyecto Western Europe:**

$r_{WE} = \frac{35 + 80 + 130 + 160 + 175}{5 \times 250} \approx 0,464 \, (46,4\%)$

**Proyecto Southeast U.S.:**

$r_{SE} = \frac{18 + 22 + 25 + 30 + 32}{5 \times 50} \approx 0,508 \, (50,8\%)$

El proyecto Southeast U.S. tiene una mayor rentabilidad anual promedio de los flujos de caja por euro invertido.

### Ventajas y Desventajas del Método de Rentabilidad Anual Promedio de los FC

**Ventajas:**
- Fácil de calcular y de interpretar.
- Es una tasa anual.
- Se basa en los flujos de caja.

**Desventajas:**
- No tiene en cuenta el valor del dinero en el tiempo.
- No tiene en cuenta el riesgo.

## Métodos Dinámicos de Decisión de Inversiones

### Plazo de Recuperación Descontado

El plazo de recuperación descontado tiene en cuenta el valor del dinero en el tiempo.

- Utiliza un tipo de interés para actualizar los flujos de caja.
- Sigue ignorando los flujos de caja después del plazo arbitrario.

**Suponiendo una tasa de actualización del 18\%:**

**Proyecto Western Europe**

$\begin{array}{|c|c|c|c|}\hline\text{Año} & \text{FC} & \text{VA(FC)} & \text{Acum VA(FC)} \\\hline0 & -250 & -250 & -250 \\1 & 30 & 25,42 & -224,58 \\2 & 80 & 57,45 & -167,12\\3 & 130 & 79,12 & -88,00 \\4 & 160 & 82,53 & -5,47 \\5 & 175 & 76,49 & 71,02 \\\hline\end{array}$

**Proyecto Southeast U.S.**

$\begin{array}{|c|c|c|c|}\hline\text{Año} & \text{FC} & \text{VA(FC)} & \text{Acum VA(FC)} \\\hline0 & -50 & -50 & -50 \\1 & 18 & 15,25 & -34,75 \\2 & 22 & 15,80 & -18,95 \\3 & 25 & 15,22 & -3,73 \\4 & 30 & 15,47 & 11,74 \\5 & 32 & 13,99 & 25,73 \\\hline\end{array}$

El Proyecto Southeast U.S. tiene un menor plazo de recuperación descontado.

### Ventajas y Desventajas del Plazo de Recuperación Descontado

**Ventajas:**
- Es fácil de calcular.
- Es fácil de interpretar.
- Se basa en los flujos de caja.
- Tiene en cuenta el valor del dinero en el tiempo.

**Desventajas:**
- El plazo de recuperación es arbitrario.
- No maximiza el valor de los inversores.

## Valor Actual Neto (VAN)

El VAN es la suma de los valores actuales de los flujos de caja de un proyecto de inversión. Al calcular el valor actual de los flujos de caja se tiene en cuenta el valor del dinero en el tiempo. Con una tasa de descuento apropiada se tiene en cuenta el riesgo del proyecto.

$\text{VAN} = \sum_{t=0}^{N} \frac{FC_t}{(1 + r)^t}$

Aceptamos el proyecto si el VAN $>$ 0.

**Suponiendo una tasa de actualización del 18\% para Global Wireless, el VAN es:**

**Proyecto Western Europe:**

$\text{VAN} = -250 + \frac{35}{(1,18)^1} + \frac{80}{(1,18)^2} + \frac{130}{(1,18)^3} + \frac{160}{(1,18)^4} + \frac{175}{(1,18)^5} = \$75,3 \, \text{millones}$

**Proyecto Southeast U.S.:**

$\text{VAN} = -50 + \frac{18}{(1,18)^1} + \frac{22}{(1,18)^2} + \frac{25}{(1,18)^3} + \frac{30}{(1,18)^4} + \frac{32}{(1,18)^5} = \$25,7 \, \text{millones}$

¿Debería invertir en un proyecto o en ambos?

### Ventajas y Desventajas del VAN

**Ventajas:**
- Se basa en los flujos de caja y no en los beneficios contables.
- Toma en cuenta el valor de dinero en el tiempo.
- Diferencia el riesgo entre diferentes proyectos.

**Desventajas:**
- Le falta ser tan intuitivo como el plazo de recuperación.
- No distingue correctamente entre alternativas de inversión.

El VAN es el rey de las decisiones de inversión.

## Tasa Interna de Rentabilidad (TIR)

La TIR es la tasa de actualización que hace cero el VAN de un proyecto.

**La regla de la TIR para la toma de decisiones:**
- Si la TIR es mayor al coste del capital, aceptamos el proyecto.
- Si la TIR es menor que el coste del capital, lo rechazamos.

**Proyecto Western Europe:** TIR $(r_{WE}) = 27,8\%$

$-250 + \frac{35}{(1 + r_{WE})^1} + \frac{80}{(1 + r_{WE})^2} + \frac{130}{(1 + r_{WE})^3} + \frac{160}{(1 + r_{WE})^4} + \frac{175}{(1 + r_{WE})^5} = 0$

**Proyecto Southeast U.S.:** TIR $(r_{SE}) = 36,7\%$

$-50 + \frac{18}{(1 + r_{SE})^1} + \frac{22}{(1 + r_{SE})^2} + \frac{25}{(1 + r_{SE})^3} + \frac{30}{(1 + r_{SE})^4} + \frac{32}{(1 + r_{SE})^5} = 0$

Global Wireless aceptará proyectos con una TIR mínima del 18\%.

### Ventajas y Desventajas de la TIR

**Ventajas:**
- Tiene en cuenta el valor del dinero en el tiempo.
- Utiliza flujos netos de caja, y no beneficios.
- Tiene en cuenta todos los flujos de caja del proyecto.
- La TIR es un número intuitivo.

**Desventajas:**
- “Problemas matemáticos”: TIRs múltiples, soluciones no reales.
- Problemas de escala.
- Problemas de timing.

## Índice de Rentabilidad (IR)

El IR se calcula dividiendo el valor actual de los flujos de caja por la inversión inicial.

$IR = \frac{\sum_{t=0}^{N} \frac{FC_t}{(1 + r)^t}}{\text{Inversión inicial}}$

**Criterio:** Se aceptan proyectos con $IR > 1$, lo que equivale a que el VAN $>$ 0.

**Los dos IR $>$ 1,0, por lo que ambos proyectos son aceptables.**

**Proyecto Western Europe:** IR = 75,3 millones / 250 millones $\approx 1,3$

**Proyecto Southeast U.S.:** IR = 25,7 millones / 50 millones $\approx 1,5$

Igual que la TIR, el IR plantea el problema de escala.

## Ejercicio de Elección de Proyecto

Hay que elegir entre dos proyectos de inversión y estamos sometidos a una restricción presupuestaria.   
Si sabemos que la tasa de descuento que tenemos que aplicar es del 8\%,   
¿cuál de estos dos proyectos elegiremos?  

$\begin{array}{|c|c|c|}\hline\text{Año} & \text{Proyecto 1} & \text{Proyecto 2} \\\hline0 & -20 & -10 \\1 & 8 & 4 \\2 & 10 & 7 \\3 & 17 & 9 \\\hline\end{array}$

### Paso 1: Definir los flujos de caja para cada proyecto

```python
import numpy as np
import numpy_financial as npf

# Flujos de caja para cada proyecto
proyecto_1 = [-20, 8, 10, 17]
proyecto_2 = [-10, 4, 7, 9]
tasa_descuento = 0.08

# Función para calcular el VAN
def calcular_van(flujos, tasa):
    return npf.npv(tasa, flujos)

# Función para calcular la TIR
def calcular_tir(flujos):
    return npf.irr(flujos)

# Función para calcular el IR
def calcular_ir(flujos, tasa):
    return npf.npv(tasa, flujos[1:]) / abs(flujos[0])

# Función para calcular la TIRC
def calcular_tirc(flujos, tasa):
    tir = calcular_tir(flujos)
    return (1 + tir)**(len(flujos) - 1) - 1

# Cálculos para el Proyecto 1
van_1 = calcular_van(proyecto_1, tasa_descuento)
tir_1 = calcular_tir(proyecto_1)
ir_1 = calcular_ir(proyecto_1, tasa_descuento)
tirc_1 = calcular_tirc(proyecto_1, tasa_descuento)

# Cálculos para el Proyecto 2
van_2 = calcular_van(proyecto_2, tasa_descuento)
tir_2 = calcular_tir(proyecto_2)
ir_2 = calcular_ir(proyecto_2, tasa_descuento)
tirc_2 = calcular_tirc(proyecto_2, tasa_descuento)
```

### Paso 2: Elegir el proyecto

```python
print("Resultados para el Proyecto 1:")
print(f"VAN: {van_1:.2f}")
print(f"TIR: {tir_1:.2f}")
print(f"IR: {ir_1:.2f}")
print(f"TIRC: {tirc_1:.2f}")

print("\nResultados para el Proyecto 2:")
print(f"VAN: {van_2:.2f}")
print(f"TIR: {tir_2:.2f}")
print(f"IR: {ir_2:.2f}")
print(f"TIRC: {tirc_2:.2f}")

if van_1 > van_2:
    print("\nEl Proyecto 1 es la mejor opción basado en el VAN.")
elif van_1 < van_2:
    print("\nEl Proyecto 2 es la mejor opción basado en el VAN.")
else:
    print("\nAmbos proyectos tienen el mismo VAN. Considerar otros indicadores como la TIR, IR o TIRC.")
```
